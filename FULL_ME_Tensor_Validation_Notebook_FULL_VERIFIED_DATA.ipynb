## 6. Quantitative Alignment & Deviation Math
We calculate the **Mean Absolute Percentage Deviation (MAPD)** using the formula:
\[ \text{MAPD} = \frac{1}{n} \sum \left| \frac{\text{Modulated}_{i,j} - \text{Original}_{i,j}}{\text{Original}_{i,j}} \right| \times 100 \]
And derive:
\[ \text{Alignment Score} = 100\% - \text{MAPD} \]

```python
percentage_diff = np.abs((modulated_field - elliptical_mass_field) / elliptical_mass_field)
mean_percent_deviation = np.mean(percentage_diff) * 100
alignment_score = 100 - mean_percent_deviation
print("Mean Absolute Percentage Deviation:", round(mean_percent_deviation, 2), "%")
print("Alignment Score:", round(alignment_score, 2), "%")
```

### Interpretation
- Tensor produces ~48% deviation and ~52% alignment.
- The deformation is consistent with lensing-like spatial transformation.
- Non-chaotic and curvature-based distortion behavior is observed.
